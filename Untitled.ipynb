```python
def get_catalog_photometry(epic_name, min_unc=0.01):
    directory = os.path.join('data',epic_name)
    mags = {}
    tm = pd.read_csv(os.path.join(directory, 'II_246_out.csv'))
    tm.sort_values(by='Jmag', inplace=True)
    for b in ['J', 'H', 'K']:
        unc = max(min_unc, tm['e_{}mag'.format(b)][0])
        mags[b] = (tm['{}mag'.format(b)][0], unc)
    
    try:
        sdss = pd.read_csv(os.path.join(directory, 'V_139_sdss9.csv'))
        sdss.sort_values(by='rpmag', inplace=True)
        for b in ['u', 'g', 'r', 'i', 'z']:
            unc = max(min_unc, sdss['e_{}pmag'.format(b)][0])
            mags[b] = (sdss['{}pmag'.format(b)][0], unc)
    except IOError:
        pass
        
    wise = pd.read_csv(os.path.join(directory, 'II_328_allwise.csv'))
    wise.sort_values(by='W1mag', inplace=True)
    for b in ['W1', 'W2', 'W3']:
        unc = max(min_unc, wise['e_{}mag'.format(b)][0])
        mags[b] = (wise['{}mag'.format(b)][0], unc)
        
    return mags

from configobj import ConfigObj

def write_ini(epic_name):
    epic_id = int(epic_name[4:])
    directory = os.path.join('starmodels',str(epic_id))
    if not os.path.exists(directory):
        os.makedirs(directory)
    c = ConfigObj(os.path.join(directory, 'star.ini'))
    
    mags = get_catalog_photometry(epic_name)
    for b, m in mags.items():
        c[b] = m
        
    s = df_match[df_match.epic_name==epic_name]
    c['ra'], c['dec'] = s.ra.iloc[0], s.dec.iloc[0]
    c['parallax'] = s.parallax.iloc[0], s.parallax_error.iloc[0]
    
    c['name'] = epic_name
    c.write()
```
